```python
class Empty (Exception):
    pass

class Stack:
    '''LIFO Stack implementation using a Python list as underlying storage.'''
   
    def __init__(self):
        self._data = []      
   
    def push (self, item):
        '''Add elemente to the top of stack'''
        self._data.append(item)
       
    def pop (self):
        '''Remove and return the top element from the stack;
       
        Raise a Empty exception  if the stack is empty'''
       
        if self.is_empty():
            raise Empty ('The stack is empty')
           
        return self._data.pop()
   
    def is_empty(self):
        '''Return True if stack does not contain any elements.'''
       
        if len(self._data):
            return False
        else:
            return True
       
    def top(self):
        '''Return a reference to the top element of stack,
        without removing it;
       
        Raise Empty exception if the stack is empty'''
       
        if self.is_empty():
            raise Empty('The stack is empty')
        else:
            return self._data[-1]
         
    def __len__(self)->int:
        return len(self._data)
   
    def __str__(self)->str:
        _str = "<"
       
        if self.is_empty():
            _str += " >"
            return _str
           
        for i in range(0, len(self._data) - 1):
            _str = _str + str(self._data[i]) + ","
       
        # Last element
        _str = _str + str(self._data[-1]) + ">"
           
        return _str
   
    def __iter__(self):
        for i in range(0, len(self._data)):
            yield self._data[i]
    
        
        
def stack_order(s:Stack)->Stack:
    """introducimos una pila con enteros o caracteres y la funcion devuelve otra pila con los elementos ordenados, de forma que en el tope
    de la pila se encuentre el valor más alto(será con el que se haya hecho el último push) y en lo más bajo estará el menor elemento(primer push)"""
    pila_final=Stack()
    aux=0
    if s.is_empty: return Empty()
    pila_final.push(s.pop())
    while s.is_empty()==False:
        aux=s.pop()
        while pila_final.is_empty()==False and aux < pila_final.top():
            s.push(pila_final.pop())
        pila_final.push(aux)
    return pila_final

######-----------------
prueba=Stack()
for i in (3,1,5,2):
    prueba.push(i)

imp=stack_order(prueba)
print (imp)


s = Stack()
imp=stack_order(prueba)
print (imp)

        
        


        
    
```

```python
#######------------         EJERCICIO 4.2

def stack_max_iter(s:Stack()):
    """Para devolver el valor máximo de la pila s lo que haremos será ir sacando de uno en uno los elementos de 
    la pila, y cada vez que sacamos uno realizamos una comprobación de si el valor del tope de la pila auxiliar
    es menor o mayor que el pop que hemos hecho de la pila s. 
    Como es obvio, el priemr elemento de la pila auxiliar
    será el primer pop que hagamos de s. Cuando la pila s se quede vacía, devolveremos el valor que hay en el tope 
    de la pila auxiliar, que será el máximo que buscamos.
    """
    pila_aux=Stack()
    
    pila_aux.push(s.pop())
    
    
    while s.is_empty()==False:
        valor=s.pop()  
        
        if valor>pila_aux.top():
            
            pila_aux.push(valor)
            valor=pila_aux.top() 
            
    if pila_aux.is_empty()==True:
        raise Empty 
    else:
        return pila_aux.top()
a=Stack()
for element in [1,2,45,3,12,4,5,2,46,32]:
    a.push(element)
stack_max_iter(a)   
           
        
```
